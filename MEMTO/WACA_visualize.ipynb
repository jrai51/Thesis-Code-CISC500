```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_sensor_data(filepath, sensor_type=None):
    """
    Plots the x, y, and z values over time for a given sensor type.
    
    filepath: str
        Path to the CSV file containing the sensor data.
    sensor_type: int, optional
        Filter data for a specific sensor type (e.g., 4 for accelerometer, 10 for gyroscope).
    """
    # Read and clean data
    df = clean_data(filepath)
    if df is None or df.empty:
        print(f"No valid data to plot for {filepath}")
        return

    # Convert timestamp to datetime
#     try:
#         df['timestamp'] = pd.to_datetime(df['timestamp'], unit='ms', errors='coerce')
#         df = df.dropna(subset=['timestamp'])  # Drop rows where timestamp conversion failed
#     except Exception as e:
#         print(f"Error converting timestamps: {e}")
#         return

   

    # Filter for specific sensor type if provided
    if sensor_type is not None:
        df = df[df['sensor'] == sensor_type]
        if df.empty:
            print(f"No data found for sensor type {sensor_type}.")
            return

    # Ensure data is sorted by timestamp
    df = df.sort_values('timestamp')
    
    print(df.head())  # Debugging output
    print(df.dtypes)  # Debugging output

    # Plot data
    plt.figure(figsize=(12, 6))
    plt.plot(df['timestamp'].to_numpy(), df['x'].to_numpy(), label='X-axis', color='blue', alpha=0.7)
    plt.plot(df['timestamp'].to_numpy(), df['y'].to_numpy(), label='Y-axis', color='green', alpha=0.7)
    plt.plot(df['timestamp'].to_numpy(), df['z'].to_numpy(), label='Z-axis', color='red', alpha=0.7)


    plt.title(f"{filepath}, Sensor Data Over Time{' (Sensor Type: ' + str(sensor_type) + ')' if sensor_type else ''}")
    plt.xlabel("Timestamp")
    plt.ylabel("Sensor Values")
    plt.legend()
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()

```

```python
plot_sensor_data("data/WACA/WACA/WACA_dataset/user1_2.csv", sensor_type=4)
plot_sensor_data("data/WACA/WACA/WACA_dataset/user1_1.csv", sensor_type=4)
plot_sensor_data("data/WACA/WACA/WACA_dataset/user2_1.csv", sensor_type=4)
plot_sensor_data("data/WACA/WACA/WACA_dataset/user20_1.csv", sensor_type=4)
plot_sensor_data("data/WACA/WACA/WACA_dataset/user23_1.csv", sensor_type=4)
```
